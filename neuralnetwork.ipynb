```
import pandas as pd
import numpy as np
dataset = pd.read_csv('/content/sample_data/Iris(1).csv')
#input and output features are define
X_inputs = dataset.iloc[:,:-1]
Y_output = dataset.iloc[:,4]
#silicing of data for traing and testing
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X_inputs,Y_output,test_size=.3,random_state=2)

#tarin model on dataset from sklearn
from sklearn.neural_network import MLPClassifier
clf = MLPClassifier(hidden_layer_sizes=100, activation='relu')
clf.fit(X_train,Y_train)
Y_pred = clf.predict(X_test)

#check the accuracy of model
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
print("Accuracy:",accuracy_score(Y_test,Y_pred))
```
